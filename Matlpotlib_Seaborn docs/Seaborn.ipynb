# Seaborn - Figure and Axes- Level Functions
#### A figure level function controls the figure

```python
sns.catplot(x='species', y='petal_length', kind ='swarm', data=iris )
plt.show
```

# Axes-level functions

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
sns.swarmplot(x='species',y='petal_length', data=iris, ax=ax1)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
sns.swarmplot(x='species',y='petal_length', data=iris, ax=ax1)
ax2.scatter('petal_width', 'petal_length', data=iris)
ax2.set_ylabel('petal_length')
ax2.set_xlabel('petal_width')
plt.show
```

```python
sns.set()
```

```python
sns.set_style('ticks')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
sns.swarmplot(x='species',y='petal_length',hue='species', data=iris, ax=ax1)
ax2.scatter('petal_width', 'petal_length', data=iris)
ax2.set_ylabel('petal_length')
ax2.set_xlabel('petal_width')
plt.show
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
iris = sns.load_dataset("iris")

# Create figure
plt.figure(figsize=(8, 5))

# Swarmplot with multiple colors based on species
sns.swarmplot(x='species', y='petal_length', hue='species', data=iris, size=5)

# Show plot
plt.legend(title="Species")  # Add legend for clarity
plt.show()
```

```python
fig, ax = plt.subplots()
sns.scatterplot(x='petal_length',y='petal_width', hue = 'species', data=iris)
plt.show
```

```python
sns.pairplot(data=iris, hue="petal_width")
```

```python
g = sns.PairGrid(penguins, hue="species", corner=True)
g.map_lower(sns.kdeplot, hue=None, levels=5, color=".2")
g.map_lower(sns.scatterplot, marker="+")
g.map_diag(sns.histplot, element="step", linewidth=0, kde=True)
g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((.61, .6))
```

```python
penguins = sns.load_dataset("penguins")
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
sns.kdeplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
dowjones = sns.load_dataset("dowjones")
sns.relplot(data=dowjones, x="Date", y="Price", kind="line")
```

```python
dots = sns.load_dataset("dots").query("align == 'dots'")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate",
    hue="coherence", style="choice",
)
```

```python
diamonds = sns.load_dataset("diamonds")
sns.catplot(
    data=diamonds.sort_values("color"),
    x="color", y="price", kind="boxen", hue='color'
)
```

```python
titanic = sns.load_dataset("titanic")
sns.catplot(data=titanic, x="sex", y="survived", hue="class", kind="bar")
```

## BOX & WHISKER CHART
#### COMPARE SEVERAL DISTRIBUTIONS AGAINST EACH OTHER 

```python
iris = sns.load_dataset('iris')

fig, ax = plt.subplots()
sns.boxplot('sepal_length', data=iris, ax=ax)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

fig, ax = plt.subplots()

sns.boxplot(x='sepal_length', data=iris, ax=ax)  # Use 'x' to specify the column

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.boxplot(data=iris,ax=ax);
plt.legend()
plt.show()
```
