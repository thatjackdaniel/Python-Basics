```python
from sklearn import datasets

iris_dataset = datasets.load_iris()
print(iris_dataset.keys())
```

```python
print(iris_dataset.feature_names)
```

```python
import pandas as pd

df = pd.DataFrame(iris_dataset.data, columns=iris_dataset.feature_names)
print(df.head())
```

```python
print(iris_dataset.target_names)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df[iris_dataset.feature_names[0]], df[iris_dataset.feature_names[1]], c=iris_dataset.target, cmap='viridis')
plt.xlabel(iris_dataset.feature_names[0])
plt.ylabel(iris_dataset.feature_names[1])
plt.title("Iris Dataset Scatter Plot")
plt.show()
```

# Assistant
The IRIS dataset is one of the most famous datasets in the field of machine learning and statistics. It was introduced by the British biologist and statistician Ronald A. Fisher in 1936 as an example of discriminant analysis. Heres a detailed introduction to the IRIS dataset:

### Overview
- **Purpose**: The dataset is primarily used for classification tasks and is often used as a beginner's dataset for machine learning algorithms.
- **Size**: The dataset contains 150 samples, with 50 samples for each of the three species of iris flowers.

### Features
The dataset consists of four features (attributes) measured for each sample:
1. **Sepal Length**: Length of the sepal in centimeters.
2. **Sepal Width**: Width of the sepal in centimeters.
3. **Petal Length**: Length of the petal in centimeters.
4. **Petal Width**: Width of the petal in centimeters.

### Classes
The dataset includes three classes (species) of iris flowers:
1. **Iris Setosa**
2. **Iris Versicolor**
3. **Iris Virginica**

### Usage
The IRIS dataset is commonly used for:
- **Data Visualization**: To visualize the distribution of different species based on the features.
- **Classification Algorithms**: To test and compare various classification algorithms such as k-Nearest Neighbors (k-NN), Decision Trees, Support Vector Machines (SVM), etc.
- **Dimensionality Reduction**: Techniques like PCA (Principal Component Analysis) can be applied to visualize the data in lower dimensions.

### Python Packages
To work with the IRIS dataset in Python, you can use the following packages:
- **Pandas**: For data manipulation and analysis.
- **NumPy**: For numerical operations.
- **Matplotlib** and **Seaborn**: For data visualization.
- **Scikit-learn**: For implementing machine learning algorithms and accessing the dataset directly.

You can load the IRIS dataset easily using Scikit-learn, which provides a built-in function to access it.
