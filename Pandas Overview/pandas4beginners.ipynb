```python
def has_ten_digit_number(df):
    """Checks if any column in a DataFrame contains a 10-digit number."""

    for col in df.columns:
        if df[col].astype(str).str.match(r'^\d{10}$').any():
            return True
    return False

# Example usage
df = pd.DataFrame({'A': [1234567890, 123], 'B': ['abc', 'def']})
print(has_ten_digit_number(df)) 
```
